## Math Functions

Let's use some functions from the `lib.mathutils` module:

```python
def fibonacci(n):
    """Calculate the nth Fibonacci number."""
    if n <= 0: return 0
    elif n == 1: return 1
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

for i in range(10):
    print(f"fib({i}) = {fibonacci(i)}")
```

## 📊 Live Plotting with Matplotlib

Run `uv sync` in the terminal first to install matplotlib, then run the cell below!

```python
import matplotlib.pyplot as plt
import math

x = [i * 0.1 for i in range(80)]
y1 = [math.sin(v) for v in x]
y2 = [math.cos(v) for v in x]

plt.figure(figsize=(8, 4))
plt.plot(x, y1, color='#4fc3f7', linewidth=2, label='sin(x)')
plt.plot(x, y2, color='#f48fb1', linewidth=2, label='cos(x)', linestyle='--')
plt.fill_between(x, y1, alpha=0.1, color='#4fc3f7')
plt.title('Sine & Cosine', fontsize=14, fontweight='bold', color='white')
plt.legend(facecolor='#2d2d2d', edgecolor='#444', labelcolor='white')
plt.grid(True, alpha=0.2)
plt.gca().set_facecolor('#1e1e1e')
plt.gca().tick_params(colors='#888')
plt.tight_layout()
plt.show()
print('Chart rendered above! ✅')
```
