## Math Functions

Let's use some functions from the `lib.mathutils` module:

```python
def fibonacci(n):
    """Calculate the nth Fibonacci number."""
    if n <= 0: return 0
    elif n == 1: return 1
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

for i in range(10):
    print(f"fib({i}) = {fibonacci(i)}")
```
