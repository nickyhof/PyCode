# 🐼 Pandas Data Analysis

Interactive data exploration with **pandas** — run `uv sync` in the terminal first to install dependencies.

```python
import pandas as pd

# Create a sample dataset
df = pd.DataFrame({
    'Product': ['Widget A', 'Widget B', 'Gadget X', 'Gadget Y', 'Gizmo Z'],
    'Category': ['Widgets', 'Widgets', 'Gadgets', 'Gadgets', 'Gizmos'],
    'Price': [29.99, 49.99, 79.99, 99.99, 149.99],
    'Units Sold': [500, 350, 200, 150, 80],
    'Rating': [4.5, 4.2, 4.8, 4.1, 4.9],
})

df['Revenue'] = (df['Price'] * df['Units Sold']).round(2)
print(df.to_string(index=False))
```

## Summary Statistics

```python
print(df[['Price', 'Units Sold', 'Revenue', 'Rating']].describe().round(2).to_string())
```

## Revenue by Category

```python
by_cat = df.groupby('Category').agg({
    'Revenue': 'sum',
    'Units Sold': 'sum',
    'Rating': 'mean'
}).round(2)

by_cat['Avg Price'] = (by_cat['Revenue'] / by_cat['Units Sold']).round(2)
print(by_cat.to_string())
```

## 📊 Visualize with Matplotlib

Combine pandas + matplotlib for inline charts:

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Revenue bar chart
colors = ['#4fc3f7', '#4fc3f7', '#f48fb1', '#f48fb1', '#ce93d8']
ax1.barh(df['Product'], df['Revenue'], color=colors)
ax1.set_xlabel('Revenue ($)', color='#aaa')
ax1.set_title('Revenue by Product', fontsize=13, fontweight='bold', color='white')
ax1.set_facecolor('#1e1e1e')
ax1.tick_params(colors='#888')
ax1.invert_yaxis()

# Rating scatter
scatter = ax2.scatter(df['Price'], df['Rating'], s=df['Units Sold'] / 2,
                      c=df['Revenue'], cmap='cool', alpha=0.8, edgecolors='white')
ax2.set_xlabel('Price ($)', color='#aaa')
ax2.set_ylabel('Rating', color='#aaa')
ax2.set_title('Price vs Rating (size = units)', fontsize=13, fontweight='bold', color='white')
ax2.set_facecolor('#1e1e1e')
ax2.tick_params(colors='#888')
plt.colorbar(scatter, ax=ax2, label='Revenue')

plt.tight_layout()
plt.show()
print('Charts rendered! ✅')
```
